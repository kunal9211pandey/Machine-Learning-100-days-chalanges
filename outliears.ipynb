```python
# remove outliears
q3=df['area'].quantile(0.75)
q3
```

```python
q1=df['area'].quantile(0.25)
```

```python
q1
```

```python
iqr=q3-q1
```

```python
iqr
```

```python
#  check upper and lower value
ul=q3+(1.5*iqr)
ll=q1-(1.5*iqr)
```

```python
ul
```

```python
ll
```

```python
# how to handel outlier
import numpy as np
```

```python
upper=np.where(df['area']>=ul)
lower=np.where(df['area']<=ll)
```

```python
df.drop(upper[0],inplace=True)
df.drop(lower[0],inplace=True)
```

```python
sns.boxplot(data=df,x='area')
plt.show()
```

```python
sns.displot(data=df,x='area',bins=10)
plt.show()
```

```python
sns.displot(data=df,x='area',kde=True)
plt.show()
```
