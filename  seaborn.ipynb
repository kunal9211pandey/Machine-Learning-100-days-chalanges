# Disploat

```python
sns.displot(data=mpg , x='cylinders' , bins=5)
plt.xlabel("no  of cylinders ")
plt.show()
```

```python
#kde is a smooth line that is represent overall distribuation of data , it show the density level
sns.displot(data=mpg , x='cylinders' , bins=5, kde=True)
plt.xlabel("no  of cylinders ")
plt.show()
```

```python
# kde plot without histogram
sns.displot(data=mpg , x='cylinders' , kind='kde')
plt.xlabel("no  of cylinders ")
plt.show()
```

```python
# proportional descities , through ECDF( Empirical cumulative distribution funcation )
sns.displot(data=mpg , x='cylinders' , kind='ecdf')
plt.xlabel("no  of cylinders ")
plt.show()
```

```python
# relation between two variable
sns.displot(data=mpg, x='cylinders' , y='mpg')
plt.xlabel("no of cylinders")
plt.ylabel("Mileage")
plt.show()
```

```python
# using kde
sns.displot(data=mpg, x='cylinders' , y='mpg' , kind='kde')
plt.xlabel("no of cylinders")
plt.ylabel("Mileage")
plt.show()
```

# count plot

```python
# count plot can be through of as a histogram  across value of a categorical variable(limit like in gender male and female).
sns.countplot(x='cylinders' , data=mpg)  # use diffrent colurs
plt.xlabel("no of Cylinders")
plt.ylabel("no of cars ")
plt.show()
```

```python
mpg['origin'].unique()
```

```python
# diffrent color of each country
sns.countplot(x='cylinders' , data=mpg ,hue='origin')
plt.xlabel("no of Cylinders")
plt.ylabel("no of cars ")
plt.show()
```
