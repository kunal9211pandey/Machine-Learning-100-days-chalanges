```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a random dataset of 1000 floating-point numbers between 0 and 1
dataset = np.random.rand(1000)

# Determine the range and number of bins
range_min = np.min(dataset)
range_max = np.max(dataset)
num_bins = 10

# Calculate the bin width
bin_width = (range_max - range_min) / num_bins

# Initialize an array to hold the count for each bin
bin_counts = np.zeros(num_bins)

# Count the frequency of values in each bin
for data_point in dataset:
    bin_index = int((data_point - range_min) // bin_width)
    bin_counts[bin_index] += 1

# Plot the histogram
plt.bar(range_min + np.arange(num_bins) * bin_width, bin_counts, width=bin_width)
plt.xlabel('Value Range')
plt.ylabel('Frequency')
plt.title('Histogram of Random Dataset')
plt.show()
```
